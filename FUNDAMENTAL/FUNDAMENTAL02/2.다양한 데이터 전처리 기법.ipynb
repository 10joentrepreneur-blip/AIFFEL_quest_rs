```python
# Q. ÏÇ¨Î∂ÑÏúÑ Î≤îÏúÑÏàòÎ•º Ïù¥Ïö©Ìï¥ÏÑú Ïù¥ÏÉÅÏπòÎ•º Ï∞æÎäî outlier2() Ìï®ÏàòÎ•º Íµ¨ÌòÑÌï¥Î≥¥ÏÑ∏Ïöî.
def outlier2(df, col):
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    return df[(df[col] < Q1 - 1.5 * IQR) | (df[col] > 1.5 * IQR)]

outlier2(trade, 'Î¨¥Ïó≠ÏàòÏßÄ')
```
