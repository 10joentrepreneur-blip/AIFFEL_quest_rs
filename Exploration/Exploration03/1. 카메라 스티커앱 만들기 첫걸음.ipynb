```python
# ÌïÑÏöîÌïú Ìå®ÌÇ§ÏßÄ import ÌïòÍ∏∞
import os # ÌôòÍ≤Ω Î≥ÄÏàòÎÇò ÎîîÎ†âÌÑ∞Î¶¨, ÌååÏùº Îì±Ïùò OS ÏûêÏõêÏùÑ Ï†úÏñ¥Ìï† Ïàò ÏûàÍ≤å Ìï¥Ï£ºÎäî Î™®Îìà
import cv2 # OpenCVÎùºÏù¥Î∏åÎü¨Î¶¨ ‚Üí Ïª¥Ìì®ÌÑ∞ ÎπÑÏ†Ñ Í¥ÄÎ†® ÌîÑÎ°úÍ∑∏ÎûòÎ∞çÏùÑ ÏâΩÍ≤å Ìï† Ïàò ÏûàÎèÑÎ°ù ÎèÑÏôÄÏ£ºÎäî ÎùºÏù¥Î∏åÎü¨Î¶¨
import matplotlib.pyplot as plt # Îã§ÏñëÌïú Îç∞Ïù¥ÌÑ∞Î•º ÎßéÏùÄ Î∞©Î≤ïÏúºÎ°ú ÎèÑÏãùÌôî Ìï† Ïàò ÏûàÎèÑÎ°ù ÌïòÎäî ÎùºÏù¥Î∏åÎü¨Î¶¨
import numpy as np # Îã§Ï∞®Ïõê Î∞∞Ïó¥ÏùÑ ÏâΩÍ≤å Ï≤òÎ¶¨ÌïòÍ≥† Ìö®Ïú®Ï†ÅÏúºÎ°ú ÏÇ¨Ïö©Ìï† Ïàò ÏûàÎèÑÎ°ù ÌïòÎäî ÎùºÏù¥Î∏åÎü¨Î¶¨
import dlib # Ïù¥ÎØ∏ÏßÄ Ï≤òÎ¶¨ Î∞è Í∏∞Í≥Ñ ÌïôÏäµ, ÏñºÍµ¥Ïù∏Ïãù Îì±ÏùÑ Ìï† Ïàò ÏûàÎäî c++ Î°ú Í∞úÎ∞úÎêú Í≥†ÏÑ±Îä•Ïùò ÎùºÏù¥Î∏åÎü¨Î¶¨
print("üå´üõ∏")
```

```python
home_dir = os.getenv('HOME')
my_image_path = os.path.join(home_dir, '/content/work/camera_sticker/images/image.png') # Î≥∏Ïù∏ Ïù¥ÎØ∏ÏßÄÍ∞Ä ÏûàÎäî Í≤ΩÎ°úÎ•º Í∞ÄÏßÄÍ≥† Ïò®Îã§.
img_bgr = cv2.imread(my_image_path)    # OpenCVÎ°ú Ïù¥ÎØ∏ÏßÄÎ•º Î∂àÎü¨ÏòµÎãàÎã§
img_show = img_bgr.copy()      # Ï∂úÎ†•Ïö© Ïù¥ÎØ∏ÏßÄÎ•º Îî∞Î°ú Î≥¥Í¥ÄÌï©ÎãàÎã§
plt.imshow(img_bgr) # Ïù¥ÎØ∏ÏßÄÎ•º Ï∂úÎ†•ÌïòÍ∏∞ ÏúÑÌï¥ Ï∂úÎ†•Ìï† Ïù¥ÎØ∏ÏßÄÎ•º Ïò¨Î†§Ï§ÄÎã§. (Ïã§Ï†ú Ï∂úÎ†•ÏùÄ ÌïòÏßÄ ÏïäÏùå)
plt.show() # Ïù¥ÎØ∏ÏßÄÎ•º Ï∂úÎ†•Ìï¥Ï§ÄÎã§. (Ïã§Ï†ú Ï∂úÎ†•)
```

```python
# plt.imshow Ïù¥Ï†ÑÏóê RGB Ïù¥ÎØ∏ÏßÄÎ°ú Î∞îÍæ∏Îäî Í≤ÉÏùÑ ÏûäÏßÄÎßàÏÑ∏Ïöî.
img_rgb = cv2.cvtColor(img_bgr, cv2.COLOR_BGR2RGB)
plt.imshow(img_rgb) # Ïù¥ÎØ∏ÏßÄ Ïã§Ï†ú Ï∂úÎ†•ÌïòÍ∏∞ ÏúÑÌï¥ Ïù¥ÎØ∏ÏßÄÎ•º Ïò¨Î†§Ï§å (ÏõêÌïòÎäî ÏÇ¨Ïù¥Ï¶àÏùò ÌîΩÏÖÄÏùÑ ÏõêÌïòÎäî ÏÉâÏúºÎ°ú Ï±ÑÏõåÏÑú ÎßåÎì† Í∑∏Î¶ºÏù¥ÎùºÍ≥† Ìï©ÎãàÎã§.)
plt.show() # Ïù¥ÎØ∏ÏßÄ Ïã§Ï†ú Ï∂úÎ†•
```

```python
# detectorÎ•º ÏÑ†Ïñ∏Ìï©ÎãàÎã§
detector_hog = dlib.get_frontal_face_detector() # Í∏∞Î≥∏ ÏñºÍµ¥ Í∞êÏßÄÍ∏∞Î•º Î∞òÌôò
print("üå´üõ∏")
```

```python
img_rgb = cv2.cvtColor(img_bgr, cv2.COLOR_BGR2RGB)
dlib_rects = detector_hog(img_rgb, 1)   # (image, num of image pyramid)
print("üå´üõ∏")
```

```python
# Ï∞æÏùÄ ÏñºÍµ¥ ÏòÅÏó≠ Î∞ïÏä§ Î¶¨Ïä§Ìä∏
# Ïó¨Îü¨ ÏñºÍµ¥Ïù¥ ÏûàÏùÑ Ïàò ÏûàÏäµÎãàÎã§
print(dlib_rects)

for dlib_rect in dlib_rects: # Ï∞æÏùÄ ÏñºÍµ¥ ÏòÅÏó≠Ïùò Ï¢åÌëú
    l = dlib_rect.left() # ÏôºÏ™Ω
    t = dlib_rect.top() # ÏúÑÏ™Ω
    r = dlib_rect.right() # Ïò§Î•∏Ï™Ω
    b = dlib_rect.bottom() # ÏïÑÎûòÏ™Ω

    cv2.rectangle(img_show, (l,t), (r,b), (0,255,0), 2, lineType=cv2.LINE_AA) # ÏãúÏûëÏ†êÏùò Ï¢åÌëúÏôÄ Ï¢ÖÎ£åÏ†ê Ï¢åÌëúÎ°ú ÏßÅÍ∞Å ÏÇ¨Í∞ÅÌòïÏùÑ Í∑∏Î¶º

img_show_rgb =  cv2.cvtColor(img_show, cv2.COLOR_BGR2RGB)
plt.imshow(img_show_rgb)
plt.show()
```

```python
model_path = os.path.join(home_dir, 'work/camera_sticker/models/shape_predictor_68_face_landmarks.dat')
    # Ï†ÄÏû•Ìïú landmark Î™®Îç∏Ïùò Ï£ºÏÜåÎ•º model_path Î≥ÄÏàòÏóê Ï†ÄÏû•
landmark_predictor = dlib.shape_predictor(model_path)
    # dlib ÎùºÏù¥Î∏åÎü¨Î¶¨Ïùò shape_predictor Ìï®ÏàòÎ•º Ïù¥Ïö©ÌïòÏó¨ Î™®Îç∏ÏùÑ Î∂àÎü¨Ïò¥
    # landmark_predictorÎäî RGBÏù¥ÎØ∏ÏßÄÏôÄ dlib.rectangleÎ•º ÏûÖÎ†• Î∞õÍ≥† dlib.full_object_detectionÎ•º Î∞òÌôò
    # dlib.rectangle: ÎÇ¥Î∂ÄÎ•º ÏòàÏ∏°ÌïòÎäî Î∞ïÏä§
    # dlib.full_object_detection: Í∞Å Íµ¨ÏÑ± ÏöîÏÜåÏùò ÏúÑÏπòÏôÄ, Ïù¥ÎØ∏ÏßÄ ÏÉÅÏùò Í∞ùÏ≤¥Ïùò ÏúÑÏπòÎ•º ÎÇòÌÉÄÎÉÑ
print("üå´üõ∏")
```

```python
list_landmarks = []
    # ÎûúÎìúÎßàÌÅ¨Ïùò ÏúÑÏπòÎ•º Ï†ÄÏû•Ìï† list ÏÉùÏÑ±

# ÏñºÍµ¥ ÏòÅÏó≠ Î∞ïÏä§ ÎßàÎã§ face landmarkÎ•º Ï∞æÏïÑÎÉÖÎãàÎã§
# face landmark Ï¢åÌëúÎ•º Ï†ÄÏû•Ìï¥Îë°ÎãàÎã§
for dlib_rect in dlib_rects:
    points = landmark_predictor(img_rgb, dlib_rect)
        # Î™®Îì† landmarkÏùò ÏúÑÏπòÏ†ïÎ≥¥Î•º points Î≥ÄÏàòÏóê Ï†ÄÏû•
    list_points = list(map(lambda p: (p.x, p.y), points.parts()))
        # Í∞ÅÍ∞ÅÏùò landmark ÏúÑÏπòÏ†ïÎ≥¥Î•º (x,y) ÌòïÌÉúÎ°ú Î≥ÄÌôòÌïòÏó¨ list_points Î¶¨Ïä§Ìä∏Î°ú Ï†ÄÏû•
    list_landmarks.append(list_points)
        # list_landmarksÏóê ÎûúÎìúÎßàÌÅ¨ Î¶¨Ïä§Ìä∏Î•º Ï†ÄÏû•

print(len(list_landmarks[0]))
    # ÏñºÍµ¥Ïù¥ nÍ∞úÏù∏ Í≤ΩÏö∞ list_landmarksÎäî nÍ∞úÏùò ÏõêÏÜåÎ•º Í∞ñÍ≥†
    # Í∞Å ÏõêÏÜåÎäî 68Í∞úÏùò ÎûúÎìúÎßàÌÅ¨ ÏúÑÏπòÍ∞Ä ÎÇòÏó¥Îêú list
    # list_landmarksÏùò ÏõêÏÜåÍ∞Ä 1Í∞úÏù¥ÎØÄÎ°ú list_landmarks[1]ÏùÑ Ìò∏Ï∂úÌïòÎ©¥ IndexErrorÍ∞Ä Î∞úÏÉù
```

```python
# list_landmarksÏùò ÏõêÏÜåÍ∞Ä 1Í∞úÏù¥ÎØÄÎ°ú ÏïÑÎûò Î∞òÎ≥µÎ¨∏ÏùÄ ÌïúÎ≤àÎßå Ïã§ÌñâÎê®
for landmark in list_landmarks:
    for point in landmark:
        cv2.circle(img_show, point, 1, (0, 255, 255), -1)
            # cv2.circle: OpenCVÏùò ÏõêÏùÑ Í∑∏Î¶¨Îäî Ìï®Ïàò
            # img_show Ïù¥ÎØ∏ÏßÄ ÏúÑ Í∞ÅÍ∞ÅÏùò pointÏóê
            # ÌÅ¨Í∏∞Í∞Ä 2Ïù¥Í≥† (0, 255, 255)ÏÉâÏúºÎ°ú ÎÇ¥Î∂ÄÍ∞Ä Ï±ÑÏõåÏßÑ(-1) ÏõêÏùÑ Í∑∏Î¶º
            # (ÎßàÏßÄÎßâ Ïù∏ÏàòÍ∞Ä ÏûêÏó∞ÏàòÎùºÎ©¥ Í∑∏ÎßåÌÅºÏùò ÎëêÍªòÏùò ÏÑ†ÏúºÎ°ú ÏõêÏù¥ Í∑∏Î†§Ïßê)

img_show_rgb = cv2.cvtColor(img_show, cv2.COLOR_BGR2RGB)
    # RGB Ïù¥ÎØ∏ÏßÄÎ°ú Ï†ÑÌôò
plt.imshow(img_show_rgb)
    # Ïù¥ÎØ∏ÏßÄÎ•º Ï§ÄÎπÑ
plt.show()
    # Ïù¥ÎØ∏ÏßÄÎ•º Ï∂úÎ†•
```

```python

# zip() : Îëê Í∑∏Î£πÏùò Îç∞Ïù¥ÌÑ∞Î•º ÏÑúÎ°ú ÏóÆÏñ¥Ï£ºÎäî ÌååÏù¥Ïç¨Ïùò ÎÇ¥Ïû• Ìï®Ïàò
# dlib_rectsÏôÄ list_landmarks Îç∞Ïù¥ÌÑ∞Î•º ÏóÆÏñ¥ Ï£ºÏóàÏùå
# dlib_rects : ÏñºÍµ¥ ÏòÅÏó≠ÏùÑ Ï†ÄÏû•ÌïòÍ≥† ÏûàÎäî Í∞í
# ‚Üí rectangles[[(345, 98) (531, 284)]]
# list_landmarks : 68Í∞úÏùò ÎûúÎìúÎßàÌÅ¨ Í∞í Ï†ÄÏû•(Ïù¥Î™©Íµ¨ÎπÑ ÏúÑÏπò(x,y))
# ‚Üí [[(368, 153), (368, 174), (371, 195), (376, 215), (382, 235), (391, 252), (404, 266), (420, 277), (441, 279), (461, 276), (480, 266), (495, 252), (508, 235), (516, 216), (520, 195), (523, 173), (524, 151), (377, 131), (386, 119), (399, 116), (413, 118), (425, 124), (454, 121), (466, 114), (481, 112), (494, 114), (503, 127), (439, 143), (438, 156), (438, 169), (437, 182), (423, 197), (431, 199), (438, 200), (446, 199), (454, 197), (391, 147), (399, 142), (409, 143), (416, 149), (408, 150), (399, 150), (464, 147), (472, 141), (482, 141), (489, 145), (482, 149), (473, 149), (411, 227), (421, 222), (432, 218), (439, 220), (446, 218), (458, 222), (471, 227), (458, 234), (446, 238), (438, 238), (431, 238), (420, 235), (415, 227), (432, 227), (439, 228), (447, 227), (466, 227), (446, 228), (438, 229), (431, 228)]]

for dlib_rect, landmark in zip(dlib_rects, list_landmarks): # ÏñºÍµ¥ ÏòÅÏó≠ÏùÑ Ï†ÄÏû•ÌïòÍ≥† ÏûàÎäî Í∞íÍ≥º 68Í∞úÏùò ÎûúÎìúÎßàÌÅ¨Î•º Ï†ÄÏû•ÌïòÍ≥† ÏûàÎäî Í∞íÏúºÎ°ú Î∞òÎ≥µÎ¨∏ Ïã§Ìñâ
    print (landmark[30]) # ÏΩîÏùò indexÎäî 30 ÏûÖÎãàÎã§
    x = landmark[30][0] # Ïù¥ÎØ∏ÏßÄÏóêÏÑú ÏΩî Î∂ÄÏúÑÏùò xÍ∞í
    y = landmark[30][1] - dlib_rect.height()//2 # Ïù¥ÎØ∏ÏßÄÏóêÏÑú ÏΩî Î∂ÄÏúÑÏùò yÍ∞í - ÏñºÍµ¥ ÏòÅÏó≠Ïùò ÏÑ∏Î°úÎ•º Ï∞®ÏßÄÌïòÎäî ÌîΩÏÖÄÏùò Ïàò//2 ‚Üí (437, 182-(186+1//2))
    w = h = dlib_rect.width() # ÏñºÍµ¥ ÏòÅÏó≠Ïùò Í∞ÄÎ°úÎ•º Ï∞®ÏßÄÌïòÎäî ÌîΩÏÖÄÏùò Ïàò (531-345+1) ‚Üí max(x) - min(x) +1(ÌîΩÏÖÄÏùò Ïàò Ïù¥Í∏∞ ÎïåÎ¨∏Ïóê 1ÏùÑ ÎçîÌï¥Ï§å ‚Üí ÌîΩÏÖÄ ÏàòÎäî Ï†ê ÌïòÎÇòÎ°úÎèÑ 1Ïù¥ Îê®)
    print (f'(x,y) : ({x},{y})')
    print (f'(w,h) : ({w},{h})')
```

```python
sticker_path = os.path.join(home_dir, '/content/work/camera_sticker/images/cat-whiskers.png') # Í≥†ÏñëÏù¥ ÏàòÏóº Ïù¥ÎØ∏ÏßÄÏùò Í≤ΩÎ°ú
img_sticker = cv2.imread(sticker_path) # Ïä§Ìã∞Ïª§ Ïù¥ÎØ∏ÏßÄÎ•º Î∂àÎü¨ÏòµÎãàÎã§ // cv2.imread(Ïù¥ÎØ∏ÏßÄ Í≤ΩÎ°ú) ‚Üí imageÍ∞ùÏ≤¥ ÌñâÎ†¨ÏùÑ Î∞òÌôò
img_sticker = cv2.resize(img_sticker, (w,h)) # Ïä§Ìã∞Ïª§ Ïù¥ÎØ∏ÏßÄ Ï°∞Ï†ï ‚Üí w,hÎäî ÏñºÍµ¥ ÏòÅÏó≠Ïùò Í∞ÄÎ°úÎ•º Ï∞®ÏßÄÌïòÎäî ÌîΩÏÖÄÏùò Ïàò(187) // cv2.resize(imageÍ∞ùÏ≤¥ ÌñâÎ†¨, (Í∞ÄÎ°ú Í∏∏Ïù¥, ÏÑ∏Î°ú Í∏∏Ïù¥))
print (img_sticker.shape) # ÏÇ¨Ïù¥Ï¶àÎ•º Ï°∞Ï†ïÌïú Í≥†ÏñëÏù¥ ÏàòÏóº Ïù¥ÎØ∏ÏßÄÏùò Ï∞®Ïõê ÌôïÏù∏
```

```python
import matplotlib.pyplot as plt
img = cv2.imread(sticker_path, cv2.IMREAD_UNCHANGED)
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGRA2RGBA))
plt.axis('off')
plt.title("Ìà¨Î™Ö Î∞∞Í≤Ω ÌôïÏù∏Ïö© ÎØ∏Î¶¨Î≥¥Í∏∞")
plt.show()
```

```python
# x,y,w,h Î™®Îëê ÏúÑÏóêÏÑú Î∞òÎ≥µÎ¨∏ ÏïàÏóêÏÑú ÏßÄÏ†ïÌï¥Ï§Ä Í∞íÏûÑ
# xÎäî Ïù¥ÎØ∏ÏßÄÏóêÏÑú ÏΩî Î∂ÄÏúÑÏùò xÍ∞í = 437
# yÎäî Ïù¥ÎØ∏ÏßÄÏóêÏÑú ÏΩî Î∂ÄÏúÑÏùò yÍ∞í = 89
# wÎäî ÏñºÍµ¥ ÏòÅÏó≠Ïùò Í∞ÄÎ°úÎ•º Ï∞®ÏßÄÌïòÎäî ÌîΩÏÖÄÏùò Ïàò = 187
# hÎäî ÏñºÍµ¥ ÏòÅÏó≠Ïùò Í∞ÄÎ°úÎ•º Ï∞®ÏßÄÌïòÎäî ÌîΩÏÖÄÏùò Ïàò = 187
refined_x = x - w // 2 # 437 - (187//2) = 437-93 = 344
refined_y = y - h # 89-187 = -98
# ÏõêÎ≥∏ Ïù¥ÎØ∏ÏßÄÏóê Ïä§Ìã∞Ïª§ Ïù¥ÎØ∏ÏßÄÎ•º Ï∂îÍ∞ÄÌïòÍ∏∞ ÏúÑÌï¥ÏÑú x, y Ï¢åÌëúÎ•º Ï°∞Ï†ïÌï©ÎãàÎã§. Ïù¥ÎØ∏ÏßÄ ÏãúÏûëÏ†êÏùÄ top-left Ï¢åÌëúÏù¥Í∏∞ ÎïåÎ¨∏ÏûÖÎãàÎã§.
# Ï¶â, refined_x, refined_yÍ∞íÏóêÏÑú Í≥†ÏñëÏù¥ ÏàòÏóº Ïù¥ÎØ∏ÏßÄÍ∞Ä ÏãúÏûëÎê®
print (f'(x,y) : ({refined_x},{refined_y})') # ÏùåÏàò Î∞úÏÉù : Ïù¥ÎØ∏ÏßÄ Î≤îÏúÑÎ•º Î≤óÏñ¥ÎÇ®
# Ïö∞Î¶¨Îäî ÌòÑÏû¨ ÏΩî ÏûêÎ¶¨Ïóê Í≥†ÏñëÏù¥ ÏàòÏóºÏùÑ ÎëêÍ≥† Ïã∂ÏùÄÍ±¥Îç∞, Ïù¥ÎßàÏúÑÏπò - ÏôïÍ¥Ä ÎÜíÏù¥Î•º ÌñàÎçîÎãà Ïù¥ÎØ∏ÏßÄÏùò Î≤îÏúÑÎ•º Ï¥àÍ≥ºÌïòÏó¨ ÏùåÏàòÍ∞Ä ÎÇòÏò§Îäî Í≤É
# opencvÎäî ndarrayÎç∞Ïù¥ÌÑ∞Î•º ÏÇ¨Ïö©ÌïòÎäîÎç∞, ndarrayÎäî ÏùåÏàòÏù∏Îç±Ïä§Ïóê Ï†ëÍ∑º Î∂àÍ∞ÄÌïòÎØÄÎ°ú Ïä§Ìã∞Ïª§ Ïù¥ÎØ∏ÏßÄÎ•º ÏûòÎùº Ï§òÏïº ÌïúÎã§.
```

```python
# ÏôïÍ¥Ä Ïù¥ÎØ∏ÏßÄÍ∞Ä Ïù¥ÎØ∏ÏßÄ Î∞ñÏóêÏÑú ÏãúÏûëÌïòÏßÄ ÏïäÎèÑÎ°ù Ï°∞Ï†ïÏù¥ ÌïÑÏöîÌï®
# Ï¢åÌëú ÏàúÏÑúÍ∞Ä y,xÏûÑÏóê Ïú†ÏùòÌïúÎã§. (y,x,rgb channel)
# ÌòÑÏû¨ ÏÉÅÌô©ÏóêÏÑúÎäî -y ÌÅ¨Í∏∞ÎßåÌÅº Ïä§Ìã∞Ïª§Î•º crop ÌïòÍ≥†, top Ïùò xÏ¢åÌëúÏôÄ y Ï¢åÌëúÎ•º Í∞ÅÍ∞ÅÏùò Í≤ΩÏö∞Ïóê ÎßûÏ∂∞ ÏõêÎ≥∏ Ïù¥ÎØ∏ÏßÄÏùò Í≤ΩÍ≥Ñ Í∞íÏúºÎ°ú ÏàòÏ†ïÌïòÎ©¥ ÏïÑÎûòÏôÄ Í∞ôÏùÄ ÌòïÏãùÏúºÎ°ú ÎÇòÏòµÎãàÎã§.
# ÏùåÏàòÍ∞í ÎßåÌÅº ÏôïÍ¥Ä Ïù¥ÎØ∏ÏßÄ(ÌòπÏùÄ Ï∂îÌõÑ Ï†ÅÏö©Ìï† Ïä§Ìã∞Ïª§ Ïù¥ÎØ∏ÏßÄ)Î•º ÏûêÎ•∏Îã§.
if refined_x < 0:
    img_sticker = img_sticker[:, -refined_x:]
    refined_x = 0
# ÏôïÍ¥Ä Ïù¥ÎØ∏ÏßÄÎ•º ÏîåÏö∞Í∏∞ ÏúÑÌï¥ ÏôïÍ¥Ä Ïù¥ÎØ∏ÏßÄÍ∞Ä ÏãúÏûëÌï† yÏ¢åÌëú Í∞í Ï°∞Ï†ï
if refined_y < 0:
    img_sticker = img_sticker[-refined_y:, :] # refined_yÍ∞Ä -98Ïù¥ÎØÄÎ°ú, img_sticker[98: , :]Í∞Ä ÎêúÎã§. (187, 187, 3)ÏóêÏÑú (89, 187, 3)Ïù¥ Îê® (187Í∞ú Ï§ëÏóêÏÑú 98Í∞úÍ∞Ä ÏûòÎ†§ÎÇòÍ∞ê)
    refined_y = 0

print (f'(x,y) : ({refined_x},{refined_y})')
```

```python
import os, cv2, numpy as np, matplotlib.pyplot as plt

# 1Ô∏è‚É£ Í∏∞Î≥∏ Ïù¥ÎØ∏ÏßÄ Í≤ΩÎ°ú ÏÑ§Ï†ï
base = "/content/work/camera_sticker/images"
img_bgr = cv2.imread(f"{base}/image.png", cv2.IMREAD_COLOR)

if img_bgr is None:
    raise FileNotFoundError("‚ùå ÏñºÍµ¥ Ïù¥ÎØ∏ÏßÄ(image.png)Î•º Ï∞æÏùÑ Ïàò ÏóÜÏäµÎãàÎã§.")

# 2Ô∏è‚É£ Í≥†ÏñëÏù¥ ÏàòÏóº Ïä§Ìã∞Ïª§ PNG Î∂àÎü¨Ïò§Í∏∞ (Ìà¨Î™Ö Î∞∞Í≤Ω Ï∂îÏ≤ú)
sticker_path = f"{base}/cat-whiskers.png"
stk = cv2.imread(sticker_path, cv2.IMREAD_UNCHANGED)
if stk is None:
    raise FileNotFoundError(f"‚ùå Ïä§Ìã∞Ïª§ ÌååÏùºÏùÑ Ï∞æÏùÑ Ïàò ÏóÜÏäµÎãàÎã§: {sticker_path}")

# 3Ô∏è‚É£ ÏòàÏãú ÎûúÎìúÎßàÌÅ¨ (ÏßÅÏ†ë dlib/mediapipe Îì±ÏúºÎ°ú Íµ¨Ìïú 68Í∞ú ÎûúÎìúÎßàÌÅ¨ Î∞∞Ïó¥)
# landmarks[i] = (x, y)
# ÏòàÏãú: landmarks = np.array([...])
# Ïã§Ï†ú ÏΩîÎìúÏóêÏÑúÎäî Ïù¥ÎØ∏ Ï∂îÏ∂úÎêú ÎûúÎìúÎßàÌÅ¨Î•º ÏÇ¨Ïö©ÌïòÏÑ∏Ïöî.

# ÏòàÏãúÏö©ÏúºÎ°ú Ïù¥Ï†Ñ ÏãúÍ∞ÅÌôîÏóêÏÑú ÏçºÎçò ÎûúÎìúÎßàÌÅ¨ Î∞∞Ïó¥ÏùÑ Í∞ÄÏ†ïÌï©ÎãàÎã§.
# landmarks.shape = (68, 2)

# 4Ô∏è‚É£ ÏΩî Î∞ë(33)Í≥º ÏúóÏûÖÏà† Ï§ëÏïô(51) Í∏∞Ï§ÄÏúºÎ°ú ÏàòÏóº Ï§ëÏã¨ Í≥ÑÏÇ∞
p33 = landmarks[33]
p51 = landmarks[51]
cx = int((p33[0] + p51[0]) / 2)
cy = int((p33[1] + p51[1]) / 2)

# 5Ô∏è‚É£ ÏûÖ ÏñëÏ™Ω(48~54) Í±∞Î¶¨ Í∏∞Ï§ÄÏúºÎ°ú Ïä§Ìã∞Ïª§ ÌÅ¨Í∏∞ Í≤∞Ï†ï
mouth_left  = landmarks[48]
mouth_right = landmarks[54]
mouth_w = np.linalg.norm(mouth_right - mouth_left)
target_w = int(mouth_w * 1.6)       # ÏûÖ ÎÑàÎπÑ ÎåÄÎπÑ 1.6Î∞∞ Ìè≠
scale = target_w / stk.shape[1]
target_h = int(stk.shape[0] * scale)
stk = cv2.resize(stk, (target_w, target_h), interpolation=cv2.INTER_AREA)

# 6Ô∏è‚É£ Ïä§Ìã∞Ïª§ ÏúÑÏπò Ï°∞Ï†ï (ÏÇ¥Ïßù ÏïÑÎûòÏ™ΩÏúºÎ°ú ÎÇ¥Î¶¨Í∏∞)
dx, dy = 0, int(0.15 * target_h)
x = int(cx - target_w / 2 + dx)
y = int(cy - target_h / 2 + dy)

# 7Ô∏è‚É£ Í≤ΩÍ≥Ñ ÌÅ¥Îû®ÌîÑ + ÎèôÏùº ÌÅ¨Í∏∞ ÌÅ¨Î°≠
H, W = img_bgr.shape[:2]
sx0 = max(0, -x)
sy0 = max(0, -y)
x0, y0 = max(0, x), max(0, y)
crop_w = min(target_w - sx0, W - x0)
crop_h = min(target_h - sy0, H - y0)
if crop_w <= 0 or crop_h <= 0:
    raise ValueError("Ïä§Ìã∞Ïª§Í∞Ä ÌôîÎ©¥ Î∞ñÏûÖÎãàÎã§. ÏúÑÏπòÎ•º Ï°∞Ï†ïÌïòÏÑ∏Ïöî.")

stk_crop = stk[sy0:sy0+crop_h, sx0:sx0+crop_w]
roi = img_bgr[y0:y0+crop_h, x0:x0+crop_w]

# 8Ô∏è‚É£ Ìï©ÏÑ± (Ìà¨Î™Ö or Ìù∞ Î∞∞Í≤Ω ÏûêÎèô Ï≤òÎ¶¨)
if stk_crop.shape[2] == 4:  # BGRA
    alpha = stk_crop[:, :, 3:4].astype(np.float32) / 255.0
    bgr = stk_crop[:, :, :3].astype(np.float32)
    out = (alpha * bgr + (1 - alpha) * roi.astype(np.float32)).astype(np.uint8)
else:
    gray = cv2.cvtColor(stk_crop, cv2.COLOR_BGR2GRAY)
    _, mask = cv2.threshold(gray, 250, 255, cv2.THRESH_BINARY_INV)
    bg = cv2.bitwise_and(roi, roi, mask=cv2.bitwise_not(mask))
    fg = cv2.bitwise_and(stk_crop, stk_crop, mask=mask)
    out = cv2.add(bg, fg)

# 9Ô∏è‚É£ Ìï©ÏÑ± Í≤∞Í≥º Î∞òÏòÅ
img_bgr[y0:y0+crop_h, x0:x0+crop_w] = out

# üîü Í≤∞Í≥º Ï∂úÎ†•
plt.figure(figsize=(5,6))
plt.imshow(cv2.cvtColor(img_bgr, cv2.COLOR_BGR2RGB))
plt.axis('off')
plt.title("üê± Í≥†ÏñëÏù¥ ÏàòÏóº Ìï©ÏÑ± (ÎûúÎìúÎßàÌÅ¨ 51‚Äì33 Ï§ëÏã¨)")
plt.show()
```

```python
import os, shutil, cv2, numpy as np, matplotlib.pyplot as plt

# 0) Í≤ΩÎ°ú ÎèôÍ∏∞Ìôî: ~ (=/root) Ïóê ÏûàÏúºÎ©¥ /contentÎ°ú Î≥µÏÇ¨
src_dir_root = "/root/work/camera_sticker/images"
dst_dir = "/content/work/camera_sticker/images"
os.makedirs(dst_dir, exist_ok=True)

for name in ["image.png", "king.png", "hero.png"]:
    src = os.path.join(src_dir_root, name)
    dst = os.path.join(dst_dir, name)
    if os.path.exists(src) and not os.path.exists(dst):
        shutil.copyfile(src, dst)

# 1) Î°úÎìú (Î∞òÎìúÏãú /content Í≤ΩÎ°ú ÏÇ¨Ïö©)
base = "/content/work/camera_sticker/images"
img = cv2.imread(f"{base}/image.png", cv2.IMREAD_COLOR)
stk = cv2.imread(f"{base}/king.png", cv2.IMREAD_UNCHANGED)  # E-8-3.pngÎ•º king.pngÎ°ú Ïì¥ Í≤É

if img is None: raise FileNotFoundError("image.png ÏóÜÏùå (~/ ÎòêÎäî /content Í≤ΩÎ°ú ÌôïÏù∏)")
if stk is None: raise FileNotFoundError("king.png ÏóÜÏùå (~/ ÎòêÎäî /content Í≤ΩÎ°ú ÌôïÏù∏)")

H, W = img.shape[:2]
sh, sw = stk.shape[:2]

# 2) 1.2Î∞∞ ÌôïÎåÄ
scale = 0.7
stk = cv2.resize(stk, (int(sw*scale), int(sh*scale)), interpolation=cv2.INTER_NEAREST)
sh, sw = stk.shape[:2]

# 3) ÏÉÅÎã® Ï§ëÏïô Î∞∞Ïπò
x = (W - sw)//2
y = max(0, int(H * -0.08))  # ÏúÑÎ°ú ÏïΩÍ∞Ñ Ïò¨Î¶º (H * -0.1 ÌïòÎ©¥ Îçî ÏúÑÎ°ú Ïò¨ÎùºÍ∞ê)

# 4) Í≤ΩÍ≥Ñ ÌÅ¥Îû®ÌîÑ + ÎèôÏùº ÌÅ¨Í∏∞ ÌÅ¨Î°≠
sx0 = max(0, -x); sy0 = max(0, -y)
x0  = max(0, x);  y0  = max(0, y)
crop_w = min(sw - sx0, W - x0)
crop_h = min(sh - sy0, H - y0)
if crop_w <= 0 or crop_h <= 0:
    raise ValueError("Ïä§Ìã∞Ïª§Í∞Ä ÌôîÎ©¥ Î∞ñÏûÖÎãàÎã§. x/y Ï°∞Ï†ïÌïòÏÑ∏Ïöî.")

stk_crop = stk[sy0:sy0+crop_h, sx0:sx0+crop_w]
roi     = img[y0:y0+crop_h, x0:x0+crop_w]

# 5) Ìï©ÏÑ±
if stk_crop.shape[2] == 4:
    # ÏïåÌååÍ∞Ä ÏûàÏúºÎ©¥ ÏïåÌåå Î∏îÎ†åÎî©
    a   = (stk_crop[:,:,3:4].astype(np.float32))/255.0
    bgr = stk_crop[:,:,:3].astype(np.float32)
    out = (a*bgr + (1-a)*roi.astype(np.float32)).astype(np.uint8)
else:
    # ÏïåÌååÍ∞Ä ÏóÜÍ≥† Ìù∞ Î∞∞Í≤Ω(Ïõê)Ïù¥ ÏÑûÏó¨ ÏûàÎäî E-8-3.pngÏö©: "Î∞ùÏùÄ Ìù∞ÏÉâ"Îßå Ï†úÍ±∞
    gray = cv2.cvtColor(stk_crop, cv2.COLOR_BGR2GRAY)
    _, mask = cv2.threshold(gray, 250, 255, cv2.THRESH_BINARY_INV)  # 250‚ÜëÎ•º Ìù∞ Î∞∞Í≤ΩÏúºÎ°ú Í∞ÑÏ£º
    mask_inv = cv2.bitwise_not(mask)
    fg = cv2.bitwise_and(stk_crop, stk_crop, mask=mask)
    bg = cv2.bitwise_and(roi, roi, mask=mask_inv)
    out = cv2.add(bg, fg)

# 6) ÎçÆÏñ¥Ïì∞Í∏∞ + ÌëúÏãú
img[y0:y0+crop_h, x0:x0+crop_w] = out
plt.figure(figsize=(5,6))
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.axis('off')
plt.title("E-8-3 (king.png) 1.2Î∞∞ Ìï©ÏÑ±")
plt.show()
```
