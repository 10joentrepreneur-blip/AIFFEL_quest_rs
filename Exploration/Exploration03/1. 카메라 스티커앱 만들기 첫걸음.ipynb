```python
# 필요한 패키지 import 하기
import os # 환경 변수나 디렉터리, 파일 등의 OS 자원을 제어할 수 있게 해주는 모듈
import cv2 # OpenCV라이브러리 → 컴퓨터 비전 관련 프로그래밍을 쉽게 할 수 있도록 도와주는 라이브러리
import matplotlib.pyplot as plt # 다양한 데이터를 많은 방법으로 도식화 할 수 있도록 하는 라이브러리
import numpy as np # 다차원 배열을 쉽게 처리하고 효율적으로 사용할 수 있도록 하는 라이브러리
import dlib # 이미지 처리 및 기계 학습, 얼굴인식 등을 할 수 있는 c++ 로 개발된 고성능의 라이브러리
print("🌫🛸")
```

```python
# 찾은 얼굴 영역 박스 리스트
# 여러 얼굴이 있을 수 있습니다
print(dlib_rects)

for dlib_rect in dlib_rects: # 찾은 얼굴 영역의 좌표
    l = dlib_rect.left() # 왼쪽
    t = dlib_rect.top() # 위쪽
    r = dlib_rect.right() # 오른쪽
    b = dlib_rect.bottom() # 아래쪽

    cv2.rectangle(img_show, (l,t), (r,b), (0,255,0), 2, lineType=cv2.LINE_AA) # 시작점의 좌표와 종료점 좌표로 직각 사각형을 그림

img_show_rgb =  cv2.cvtColor(img_show, cv2.COLOR_BGR2RGB)
plt.imshow(img_show_rgb)
plt.show()
```

```python
model_path = os.path.join(home_dir, 'work/camera_sticker/models/shape_predictor_68_face_landmarks.dat')
    # 저장한 landmark 모델의 주소를 model_path 변수에 저장
landmark_predictor = dlib.shape_predictor(model_path)
    # dlib 라이브러리의 shape_predictor 함수를 이용하여 모델을 불러옴
    # landmark_predictor는 RGB이미지와 dlib.rectangle를 입력 받고 dlib.full_object_detection를 반환
    # dlib.rectangle: 내부를 예측하는 박스
    # dlib.full_object_detection: 각 구성 요소의 위치와, 이미지 상의 객체의 위치를 나타냄
print("🌫🛸")
```

```python
list_landmarks = []
    # 랜드마크의 위치를 저장할 list 생성

# 얼굴 영역 박스 마다 face landmark를 찾아냅니다
# face landmark 좌표를 저장해둡니다
for dlib_rect in dlib_rects:
    points = landmark_predictor(img_rgb, dlib_rect)
        # 모든 landmark의 위치정보를 points 변수에 저장
    list_points = list(map(lambda p: (p.x, p.y), points.parts()))
        # 각각의 landmark 위치정보를 (x,y) 형태로 변환하여 list_points 리스트로 저장
    list_landmarks.append(list_points)
        # list_landmarks에 랜드마크 리스트를 저장

print(len(list_landmarks[0]))
    # 얼굴이 n개인 경우 list_landmarks는 n개의 원소를 갖고
    # 각 원소는 68개의 랜드마크 위치가 나열된 list
    # list_landmarks의 원소가 1개이므로 list_landmarks[1]을 호출하면 IndexError가 발생
```

```python
# list_landmarks의 원소가 1개이므로 아래 반복문은 한번만 실행됨
for landmark in list_landmarks:
    for point in landmark:
        cv2.circle(img_show, point, 1, (0, 255, 255), -1)
            # cv2.circle: OpenCV의 원을 그리는 함수
            # img_show 이미지 위 각각의 point에
            # 크기가 2이고 (0, 255, 255)색으로 내부가 채워진(-1) 원을 그림
            # (마지막 인수가 자연수라면 그만큼의 두께의 선으로 원이 그려짐)

img_show_rgb = cv2.cvtColor(img_show, cv2.COLOR_BGR2RGB)
    # RGB 이미지로 전환
plt.imshow(img_show_rgb)
    # 이미지를 준비
plt.show()
    # 이미지를 출력
```

```python
sticker_path = os.path.join(home_dir, '/content/work/camera_sticker/images/cat-whiskers.png') # 고양이 수염 이미지의 경로
img_sticker = cv2.imread(sticker_path) # 스티커 이미지를 불러옵니다 // cv2.imread(이미지 경로) → image객체 행렬을 반환
img_sticker = cv2.resize(img_sticker, (w,h)) # 스티커 이미지 조정 → w,h는 얼굴 영역의 가로를 차지하는 픽셀의 수(187) // cv2.resize(image객체 행렬, (가로 길이, 세로 길이))
print (img_sticker.shape) # 사이즈를 조정한 고양이 수염 이미지의 차원 확인
```

```python
import matplotlib.pyplot as plt
img = cv2.imread(sticker_path, cv2.IMREAD_UNCHANGED)
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGRA2RGBA))
plt.axis('off')
plt.title("투명 배경 확인용 미리보기")
plt.show()
```

```python
# x,y,w,h 모두 위에서 반복문 안에서 지정해준 값임
# x는 이미지에서 코 부위의 x값 = 437
# y는 이미지에서 코 부위의 y값 = 89
# w는 얼굴 영역의 가로를 차지하는 픽셀의 수 = 187
# h는 얼굴 영역의 가로를 차지하는 픽셀의 수 = 187
refined_x = x - w // 2 # 437 - (187//2) = 437-93 = 344
refined_y = y - h # 89-187 = -98
# 원본 이미지에 스티커 이미지를 추가하기 위해서 x, y 좌표를 조정합니다. 이미지 시작점은 top-left 좌표이기 때문입니다.
# 즉, refined_x, refined_y값에서 고양이 수염 이미지가 시작됨
print (f'(x,y) : ({refined_x},{refined_y})') # 음수 발생 : 이미지 범위를 벗어남
# 우리는 현재 코 자리에 고양이 수염을 두고 싶은건데, 이마위치 - 왕관 높이를 했더니 이미지의 범위를 초과하여 음수가 나오는 것
# opencv는 ndarray데이터를 사용하는데, ndarray는 음수인덱스에 접근 불가하므로 스티커 이미지를 잘라 줘야 한다.
```

```python
# 왕관 이미지가 이미지 밖에서 시작하지 않도록 조정이 필요함
# 좌표 순서가 y,x임에 유의한다. (y,x,rgb channel)
# 현재 상황에서는 -y 크기만큼 스티커를 crop 하고, top 의 x좌표와 y 좌표를 각각의 경우에 맞춰 원본 이미지의 경계 값으로 수정하면 아래와 같은 형식으로 나옵니다.
# 음수값 만큼 왕관 이미지(혹은 추후 적용할 스티커 이미지)를 자른다.
if refined_x < 0:
    img_sticker = img_sticker[:, -refined_x:]
    refined_x = 0
# 왕관 이미지를 씌우기 위해 왕관 이미지가 시작할 y좌표 값 조정
if refined_y < 0:
    img_sticker = img_sticker[-refined_y:, :] # refined_y가 -98이므로, img_sticker[98: , :]가 된다. (187, 187, 3)에서 (89, 187, 3)이 됨 (187개 중에서 98개가 잘려나감)
    refined_y = 0

print (f'(x,y) : ({refined_x},{refined_y})')
```

```python
# 왕관 이미지가 이미지 밖에서 시작하지 않도록 조정이 필요함
# 좌표 순서가 y,x임에 유의한다. (y,x,rgb channel)
# 현재 상황에서는 -y 크기만큼 스티커를 crop 하고, top 의 x좌표와 y 좌표를 각각의 경우에 맞춰 원본 이미지의 경계 값으로 수정하면 아래와 같은 형식으로 나옵니다.
# 음수값 만큼 왕관 이미지(혹은 추후 적용할 스티커 이미지)를 자른다.
if refined_x < 0:
    img_sticker = img_sticker[:, -refined_x:]
    refined_x = 0
# 왕관 이미지를 씌우기 위해 왕관 이미지가 시작할 y좌표 값 조정
if refined_y < 0:
    img_sticker = img_sticker[-refined_y:, :] # refined_y가 -98이므로, img_sticker[98: , :]가 된다. (187, 187, 3)에서 (89, 187, 3)이 됨 (187개 중에서 98개가 잘려나감)
    refined_y = 0

print (f'(x,y) : ({refined_x},{refined_y})')
```

```python
# sticker_area는 원본이미지에서 스티커를 적용할 위치를 crop한 이미지 입니다.
# 예제에서는 (344,0) 부터 (344+187, 0+89) 범위의 이미지를 의미합니다.
# 좌표 순서가 y,x임에 유의한다. (y,x,rgb channel)
# img_show[refined_y:refined_y+img_sticker.shape[0], refined_x:refined_x+img_sticker.shape[1]]
# img_show[0:0+89, 344:344+187]
# img_show[0:89, 344:531]
# 즉, x좌표는 344~531 / y좌표는 0~89가 됨
sticker_area = img_show[refined_y:refined_y+img_sticker.shape[0], refined_x:refined_x+img_sticker.shape[1]]
# 왕관 이미지에서 사용할 부분은 0이 아닌 색이 있는 부분을 사용합니다.
# 왕관 이미지에서 0이 나오는 부분은 흰색이라는 뜻, 즉 이미지가 없다는 소리임.
# 현재 왕관 이미지에서는 왕관과 받침대 밑의 ------ 부분이 됨
# 그렇기 때문에 0인 부분(이미지가 없는 부분)은 제외하고 적용
# sticker_area는 원본 이미지에서 스티커를 적용할 위치를 미리 잘라낸 이미지입니다.
# 즉, 왕관 이미지에서 왕관 이미지가 없는 부분(왕관과 받침대 밑의 ------ 부분)은 원본 이미지에서 미리 잘라놓은 sticker_area(스티커 적용할 부분 만큼 원본 이미지에서 자른 이미지)를 적용하고,
# 나머지 부분은 스티커로 채워주면 됨
# np.where는 조건에 해당하는 인덱스만 찾아서 값을 적용하는 방법이다.
# 아래 코드에서는 img_sticker가 0일 경우(왕관 이미지에서 왕관 부분 제외한 나머지 이미지)에는 sticker_area(원본 이미지에서 스티커를 적용할 위치를 미리 잘라낸 이미지)를 적용하고,
# 나머지 부분은 img_sticker(왕관 이미지)를 적용한다.
img_show[refined_y:refined_y+img_sticker.shape[0], refined_x:refined_x+img_sticker.shape[1]] = \
    np.where(img_sticker==0,sticker_area,img_sticker).astype(np.uint8)
print("슝~")
```

```python
# 🧠 1. 설치
!pip install dlib

import cv2, dlib, numpy as np, matplotlib.pyplot as plt

# 2️⃣ 파일 경로
base = "/content/work/camera_sticker/images"
img_path = f"{base}/image.png"
sticker_path = f"{base}/cat-whiskers.png"

# 3️⃣ dlib 전처리기 & 예측기 다운로드
!wget http://dlib.net/files/shape_predictor_68_face_landmarks.dat.bz2
!bzip2 -dk shape_predictor_68_face_landmarks.dat.bz2

# 4️⃣ 얼굴 탐지 + 랜드마크 추출
detector = dlib.get_frontal_face_detector()
predictor = dlib.shape_predictor("shape_predictor_68_face_landmarks.dat")

img_bgr = cv2.imread(img_path)
gray = cv2.cvtColor(img_bgr, cv2.COLOR_BGR2GRAY)
faces = detector(gray)

if len(faces) == 0:
    raise ValueError("❌ 얼굴을 찾을 수 없습니다. 이미지 확인!")

face = faces[0]
shape = predictor(gray, face)
landmarks = np.array([[p.x, p.y] for p in shape.parts()])  # (68,2)

# --- 랜드마크 시각화 확인 ---
for (x, y) in landmarks:
    cv2.circle(img_bgr, (x, y), 2, (0, 255, 255), -1)

plt.imshow(cv2.cvtColor(img_bgr, cv2.COLOR_BGR2RGB))
plt.title("얼굴 랜드마크 (68점)")
plt.axis("off")
plt.show()
```

```python
import os, cv2, numpy as np, matplotlib.pyplot as plt

# 1️⃣ 기본 이미지 경로 설정
base = "/content/work/camera_sticker/images"
img_bgr = cv2.imread(f"{base}/image.png", cv2.IMREAD_COLOR)

if img_bgr is None:
    raise FileNotFoundError("❌ 얼굴 이미지(image.png)를 찾을 수 없습니다.")

# 2️⃣ 고양이 수염 스티커 PNG 불러오기 (투명 배경 추천)
sticker_path = f"{base}/cat-whiskers.png"
stk = cv2.imread(sticker_path, cv2.IMREAD_UNCHANGED)
if stk is None:
    raise FileNotFoundError(f"❌ 스티커 파일을 찾을 수 없습니다: {sticker_path}")

# 3️⃣ 예시 랜드마크 (직접 dlib/mediapipe 등으로 구한 68개 랜드마크 배열)
# landmarks[i] = (x, y)
# 예시: landmarks = np.array([...])
# 실제 코드에서는 이미 추출된 랜드마크를 사용하세요.

# 예시용으로 이전 시각화에서 썼던 랜드마크 배열을 가정합니다.
# landmarks.shape = (68, 2)

# 4️⃣ 코 밑(33)과 윗입술 중앙(51) 기준으로 수염 중심 계산
p33 = landmarks[33]
p51 = landmarks[51]
cx = int((p33[0] + p51[0]) / 2)
cy = int((p33[1] + p51[1]) / 2)

# 5️⃣ 입 양쪽(48~54) 거리 기준으로 스티커 크기 결정
mouth_left  = landmarks[48]
mouth_right = landmarks[54]
mouth_w = np.linalg.norm(mouth_right - mouth_left)
target_w = int(mouth_w * 1.6)       # 입 너비 대비 1.6배 폭
scale = target_w / stk.shape[1]
target_h = int(stk.shape[0] * scale)
stk = cv2.resize(stk, (target_w, target_h), interpolation=cv2.INTER_AREA)

# 6️⃣ 스티커 위치 조정 (살짝 아래쪽으로 내리기)
dx, dy = 0, int(0.15 * target_h)
x = int(cx - target_w / 2 + dx)
y = int(cy - target_h / 2 + dy)

# 7️⃣ 경계 클램프 + 동일 크기 크롭
H, W = img_bgr.shape[:2]
sx0 = max(0, -x)
sy0 = max(0, -y)
x0, y0 = max(0, x), max(0, y)
crop_w = min(target_w - sx0, W - x0)
crop_h = min(target_h - sy0, H - y0)
if crop_w <= 0 or crop_h <= 0:
    raise ValueError("스티커가 화면 밖입니다. 위치를 조정하세요.")

stk_crop = stk[sy0:sy0+crop_h, sx0:sx0+crop_w]
roi = img_bgr[y0:y0+crop_h, x0:x0+crop_w]

# 8️⃣ 합성 (투명 or 흰 배경 자동 처리)
if stk_crop.shape[2] == 4:  # BGRA
    alpha = stk_crop[:, :, 3:4].astype(np.float32) / 255.0
    bgr = stk_crop[:, :, :3].astype(np.float32)
    out = (alpha * bgr + (1 - alpha) * roi.astype(np.float32)).astype(np.uint8)
else:
    gray = cv2.cvtColor(stk_crop, cv2.COLOR_BGR2GRAY)
    _, mask = cv2.threshold(gray, 250, 255, cv2.THRESH_BINARY_INV)
    bg = cv2.bitwise_and(roi, roi, mask=cv2.bitwise_not(mask))
    fg = cv2.bitwise_and(stk_crop, stk_crop, mask=mask)
    out = cv2.add(bg, fg)

# 9️⃣ 합성 결과 반영
img_bgr[y0:y0+crop_h, x0:x0+crop_w] = out

# 🔟 결과 출력
plt.figure(figsize=(5,6))
plt.imshow(cv2.cvtColor(img_bgr, cv2.COLOR_BGR2RGB))
plt.axis('off')
plt.title("🐱 고양이 수염 합성 (랜드마크 51–33 중심)")
plt.show()
```

```python
import os, shutil, cv2, numpy as np, matplotlib.pyplot as plt

# 0) 경로 동기화: ~ (=/root) 에 있으면 /content로 복사
src_dir_root = "/root/work/camera_sticker/images"
dst_dir = "/content/work/camera_sticker/images"
os.makedirs(dst_dir, exist_ok=True)

for name in ["image.png", "king.png", "hero.png"]:
    src = os.path.join(src_dir_root, name)
    dst = os.path.join(dst_dir, name)
    if os.path.exists(src) and not os.path.exists(dst):
        shutil.copyfile(src, dst)

# 1) 로드 (반드시 /content 경로 사용)
base = "/content/work/camera_sticker/images"
img = cv2.imread(f"{base}/image.png", cv2.IMREAD_COLOR)
stk = cv2.imread(f"{base}/king.png", cv2.IMREAD_UNCHANGED)  # E-8-3.png를 king.png로 쓴 것

if img is None: raise FileNotFoundError("image.png 없음 (~/ 또는 /content 경로 확인)")
if stk is None: raise FileNotFoundError("king.png 없음 (~/ 또는 /content 경로 확인)")

H, W = img.shape[:2]
sh, sw = stk.shape[:2]

# 2) 1.2배 확대
scale = 0.7
stk = cv2.resize(stk, (int(sw*scale), int(sh*scale)), interpolation=cv2.INTER_NEAREST)
sh, sw = stk.shape[:2]

# 3) 상단 중앙 배치
x = (W - sw)//2
y = max(0, int(H * -0.08))  # 위로 약간 올림 (H * -0.1 하면 더 위로 올라감)

# 4) 경계 클램프 + 동일 크기 크롭
sx0 = max(0, -x); sy0 = max(0, -y)
x0  = max(0, x);  y0  = max(0, y)
crop_w = min(sw - sx0, W - x0)
crop_h = min(sh - sy0, H - y0)
if crop_w <= 0 or crop_h <= 0:
    raise ValueError("스티커가 화면 밖입니다. x/y 조정하세요.")

stk_crop = stk[sy0:sy0+crop_h, sx0:sx0+crop_w]
roi     = img[y0:y0+crop_h, x0:x0+crop_w]

# 5) 합성
if stk_crop.shape[2] == 4:
    # 알파가 있으면 알파 블렌딩
    a   = (stk_crop[:,:,3:4].astype(np.float32))/255.0
    bgr = stk_crop[:,:,:3].astype(np.float32)
    out = (a*bgr + (1-a)*roi.astype(np.float32)).astype(np.uint8)
else:
    # 알파가 없고 흰 배경(원)이 섞여 있는 E-8-3.png용: "밝은 흰색"만 제거
    gray = cv2.cvtColor(stk_crop, cv2.COLOR_BGR2GRAY)
    _, mask = cv2.threshold(gray, 250, 255, cv2.THRESH_BINARY_INV)  # 250↑를 흰 배경으로 간주
    mask_inv = cv2.bitwise_not(mask)
    fg = cv2.bitwise_and(stk_crop, stk_crop, mask=mask)
    bg = cv2.bitwise_and(roi, roi, mask=mask_inv)
    out = cv2.add(bg, fg)

# 6) 덮어쓰기 + 표시
img[y0:y0+crop_h, x0:x0+crop_w] = out
plt.figure(figsize=(5,6))
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.axis('off')
plt.title("E-8-3 (king.png) 1.2배 합성")
plt.show()
```
